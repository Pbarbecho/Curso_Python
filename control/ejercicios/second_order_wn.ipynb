```python
import numpy as np
import control
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})
```

```python
K=1
z = 0.2
wn=[1,2,3]

# state space model


for vwn in wn: 
  num=[K*vwn**2]
  denm=[1,2*z*vwn,vwn**2]
  model = control.tf(num,denm)
  
  t, y = control.step_response(model,20)

  # plot
  plt.plot(t,y, label=f'$w_{"n"}$={vwn}')
  plt.legend()

  plt.title("Step Response $K=1, \zeta=0.2$")
  #plt.grid()


```
