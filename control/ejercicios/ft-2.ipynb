```python
import numpy as np
import control
import matplotlib.pyplot as plt


m=200
b=80
k=2

for ck in range(10,100,20):
    num = np.array([ 1 ])
    den = np.array([ m , b, ck])
    H = control.tf(num , den)
    t, y = control.impulse_response(H)
    plt.plot(t,y, label=ck)
    plt.legend()

plt.title("Impulse Response")
plt.grid()
```
