```python
import numpy as np
import control
import matplotlib.pyplot as plt


m=20
b=4
k=10

# state space model
A=[[0,1],[-1,-3]]
B=[[0],[2]]
C=[[5,6]]
D=[[7]]
model = control.ss(A,B,C,D)


start=0
stop=50
step=0.1
t = np.arange(start,stop+1,step)

u=3
t, y, x = control.forced_response(model,t,u,return_x=True)


# plot
plt.plot(t,y, label=k)
plt.legend()

plt.title("U=x")
#plt.grid()

print(x)
plt.figure(2)
x1=x[0 ,:]
x2=x[1 ,:]
plt.plot(t,x1,t,x2)



```
