## Required Packages (Pyomo)

```python
import pyomo.environ as pyo 
```

## Create the model

```python
model = pyo.ConcreteModel()
```

## Define Desicion variables

```python
model.x1 = pyo.Var(within=pyo.NonNegativeReals)
model.x2 = pyo.Var(within=pyo.NonNegativeReals)
```

## Define Objective Function

```python
model.objective_Funct = pyo.Objective(
    expr= model.x1 + 2*model.x2)
```

## Define constraints

```python
model.constraint_1 = pyo.Constraint(expr=3*model.x1 + 4*model.x2>=1)
model.constraint_2 = pyo.Constraint(expr=2*model.x1 + 5*model.x2>=2)
```

## Solve the model

```python
opt = pyo.SolverFactory('glpk')
result = opt.solve(model)
```
