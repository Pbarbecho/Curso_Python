## Import Pyomo

```python
import pyomo.environ as pyo
```

### 1. Data

```python
I = ['E1','E2','E3','E4','E5'] # items
W = {'E1':12,'E2':2,'E3':1,'E4':4,'E5':1} # item weight
V = {'E1':4,'E2':2,'E3':1,'E4':10,'E5':2} # item values
W_max = 15
```

### 2. Model (Concrete Model)

```python
model = pyo.ConcreteModel()
```

### 3. Desicion Variables

```python
model.x = pyo.Var(I, within=pyo.Binary)
```

### 4. Objective Function

```python
model.objective_Funct = pyo.Objective(
    expr= sum(V[i]*model.x[i] for i in I),
    sense=pyo.maximize)
```

### 5. Constraints

```python
model.constraint_1 = pyo.Constraint(expr=sum(W[i]*model.x[i] for i in I) <= W_max)
```

### 6. Solve

```python
factory = pyo.SolverFactory('glpk')
factory.solve(model, tee=True)
```

```python
model.pprint()
```

## Print just results of desicion variables

```python
results = []

for i in I:
    results.append(pyo.value(model.x[i]))
print(results)
    
```
